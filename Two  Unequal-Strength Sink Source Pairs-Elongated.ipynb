The two sink source pairs are symmetrically arranged above and below the horizontal axis. 

The sources at (-8, +/- 1) are twice as strong as the sinks at (1, +/- 1).  

The contaminant flow capture by the downstream extraction wells is clearly visible.

The plume boundary can be roughly inferred from the streamline heading up/down from the stagnation point at approximately (-13,0).
