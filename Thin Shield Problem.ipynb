In this Notebook we will use the inverse Joukowski  transformation (IJT) to analyze flow at an angle  $\delta=-\frac{\pi}{3}$ around an impermeable fracture of length 8, with velocity $\frac{q_u}{A} = 5$. The fracture is represented by a horizontal line segment of length $4R=8$, centered at the origin. The IJT 
 \begin{equation}
        \xi(z)=\frac{z+(z^2-4R^2)^{1/2}}{2},
    \end{equation}
maps the line segment $[-2R,2R]$ to the circle $r=R$. Thus the image $\xi(z)$ of points belonging to the fracture $z\in[-2R,2R]$ is a circle of radius $R$ centered at the origin.
 
 Given that flow at an angle $\delta$ around a circular object of radius $R$ is modeled by a dipole (coalescence of equal strength source and sink) with potential 
  \begin{equation}
     \Omega_0(z)=e^{i\delta}[-\frac{q_u}{A}z + \frac{R^2}{z}]\, 
    \end{equation}
we find the potential $\Omega(z)$ for flow around a thin shield by forming the composition of $\Omega_0(z)$ and $\xi$. After simplifying, we obtain
\begin{equation}
    \Omega(z)= \Omega_0(\xi(z))=-\frac{q_u}{2A}(z\cos\delta - i(z^2-4R^2)^{1/2}\sin\delta).   
    \end{equation}
 
 For the specific values $R=2$, $\delta=-\frac{\pi}{3}$, and $\frac{q_u}{A} = 5$, we have
 
 \begin{equation}
     \Omega(z)=-\frac{5z}{2} + \frac{5\sqrt{3}}{2}i(z^2-16)^{1/2}. 
    \end{equation}
    
To plot ths streamlines, we let   
    
$r_1 = \sqrt{(x-4)^2+y^2},\, r_2 = \sqrt{(x+4)^2+y^2}\,$, and

$\theta_1 = \arctan(\frac{y}{x-4}),\, \theta_2 = \arctan(\frac{y}{x+4}).$

The stream function is

$\Psi=-(5/2)y+\frac{5\sqrt{3}}{2}\sqrt{(r_1r_2)}\cos(\frac{\theta_1+\theta_2}{2})$




1) Import Libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
import cmath as cm
from time import time
```

2) Plot Streamlines

```python
# create figure and axes
fig=plt.figure(figsize=(5,5))
plt.xlim(-5,5)
plt.ylim(-5,5)
# Set up Plot Grid
x = np.linspace(-10, 10, 250)
y = np.linspace(-10, 10, 250)
X, Y = np.meshgrid(x, y)
fig.suptitle('Streamlines',size=14)

#Plot the streamlines
r1 = np.sqrt((X-4)**2+Y**2)
r2 = np.sqrt((X+4)**2+Y**2)
theta1 = np.arctan2(Y,(X-4))
theta2 = np.arctan2(Y,(X+4))
#Z=(-(np.sqrt(3)/2)*Y)+(1/2)*np.sqrt(r1*r2)*np.cos((theta1+theta2)/2)
Z=(-(5/2)*Y)+np.sqrt(r1*r2)*np.cos((theta1+theta2)/2)*(5*np.sqrt(3)/2)
contours=plt.contour(X, Y, Z, levels=np.linspace(-30,30,50), colors='black');
plt.gca().clabel(contours, inline=True, fontsize=8)
#plot the fracture
plt.plot([-4,4],[0,0],color='red',linewidth=2)
plt.savefig('ThinShield.png')
```
