The Space-Leasing Plume's pollution extraction system can be modelled using a complex potential of the form
 $\Omega_0(z)=-.145 z + 3k\log(z)+2k\log(z-10.8) + 717$, with extraction strength parameter value $k=.1$.  
 
 This JNB rotates the flow by $\pi/4$ so the model is more in line with the geographic position of the SL-plume.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

<h4> OriginalSystem </h4>

```python
plt.figure(figsize=(20,15))
x = np.linspace(-15, 15, 250)
y = np.linspace(-4, 4, 250)
plt.xlim(-15, 15)
plt.ylim(-4,4)
k=.1
k0=-.145
c=720
X, Y = np.meshgrid(x, y)
#Z = k0*X+.3*k*.5*np.log(Y**2+X**2)+.2*k*.5*np.log(Y**2+(X-10.8)**2)+c
#contours=plt.contour(X, Y, Z, levels=np.linspace(700,800,600), colors='red')
Z0 = k0*Y+.3*k*np.arctan2(Y,X)+.2*k*np.arctan2(Y,(X-10.8))
contours0=plt.contour(X, Y, Z0, levels=np.linspace(-10,20,400), colors='black')
plt.gca().tick_params(labelsize=15)
#plt.clabel(contours, inline=True, fontsize=15)
plt.text(-.5,2.5,'oXX',fontsize=20,color='g')
plt.text(0,1.5,'oXX',fontsize=20,color='g')
plt.text(6,0,'oXX',fontsize=20,color='g')
plt.text(-8,1,'oXX',fontsize=20,color='g')
plt.text(-11.8,1,'oXX',fontsize=20,color='g')
plt.text(-11.8,1,'oXX',fontsize=20,color='g')
plt.text(-11.8,1,'oXX',fontsize=20,color='g')
plt.text(0,0,'oEWSL-2',fontsize=20,color='k')
plt.text(10.8,0,'oEWS-4',fontsize=20,color='k')
plt.savefig('Original.png')
```

<h4>Tilted System</h4>
The tilted system has complex potential
 $\Omega(z)=\Omega_0(e^{i\pi/4}z)=-.145 z + .3\log(e^{i\pi/4}z)++.3i(\pi/4)+.2\log((e^{i\pi/4}z-10.8)) + 717=-.145z+.3\log(z)+.2\log((e^{i\pi/4}z-10.8))+717$. Thus, the new stream function is
 
 \begin{equation}
 \Psi(z)=-.145(\frac{x}{\sqrt{2}}+\frac{y}{\sqrt{2}}) +.3\arctan(\frac{y}{x})+.3(\frac{\pi}{4}) +.2\arctan(\frac{(x+y)/\sqrt{2}}{[\frac{x}{\sqrt{2}}-\frac{y}{\sqrt{2}}-10.8]})
 \end{equation}
 

```python
plt.figure(figsize=(20,15))
x = np.linspace(-15, 15, 250)
y = np.linspace(-15, 15, 250)
plt.xlim(-15, 15)
plt.ylim(-15,15)
X, Y = np.meshgrid(x, y)
#Z = k0*X+.3*k*.5*np.log(Y**2+X**2)+.2*k*.5*np.log(Y**2+(X-10.8)**2)+c
#contours=plt.contour(X, Y, Z, levels=np.linspace(700,800,600), colors='red')
Z = -.145*(X/np.sqrt(2)+Y/np.sqrt(2))+.3*np.arctan2(Y,X)+.3*np.pi/4+.2*np.arctan2((X+Y)/np.sqrt(2),X/np.sqrt(2) - Y/np.sqrt(2)-10.8)
contours=plt.contour(X, Y, Z, levels=np.linspace(-10,20,150), colors='black')
plt.gca().tick_params(labelsize=15)
#plt.clabel(contours, inline=True, fontsize=15)

plt.savefig('Rotated.png')
```
