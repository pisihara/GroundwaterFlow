In this JNB we will plot streamlines for the complex potential 

\begin{equation}
\Omega = -z - 2\ln(z+1) + 2\ln(z-1)
\end{equation}

which corresponds flow around an impermeable boundary called a Rankine Oval.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

1) Plot of the the stream function $\Psi$ for $\Omega$.

```python
plt.figure(figsize=(20,15))
x = np.linspace(-4, 4, 250)
y = np.linspace(-3, 3, 250)
#---Edit the next cell with the correct position for the source---
plt.text(-1.05,-.05,'x',color='r',size=30)
plt.text(-1.25,-.2,'Pollution',color='r',size=35)
plt.text(-1.25,-.35,'Source',color='r',size=35)
plt.xlim(-4, 4)
plt.ylim(-2.5,2.5)
#---Edit the next cell with the correct position for the sink---
plt.text(.95,-.05,'x',color='g',size=30)
plt.text(.5,-.2,'Extraction',color='g',size=35)
plt.text(.5,-.35,'well',color='g',size=35)
plt.text(.5,-.5,'(sink)',color='g',size=35)

#---Edit the next cell with the correct position for the sink---
plt.text(.5,1.5,'o Monitoring well',color='g',size=30)

X, Y = np.meshgrid(x, y)
#---Edit the next section with the correct stream function
Z = -Y-2*np.arctan2(Y,(X+1))+2*np.arctan2(Y,(X-1))
contours=plt.contour(X, Y, Z, levels=np.linspace(-15,15,100), colors='black');
#---Plot the Dividing Streamline---------
contours=plt.contour(X, Y, Z, levels=-2*np.arctan(.0001)+2*np.arctan(.0001), colors='red');
contours=plt.contour(X, Y, Z, levels=-2*np.pi, colors='red');
contours=plt.contour(X, Y, Z, levels=2*np.pi, colors='red');
plt.text(-2.5,.5,'Dividing Streamline',color='r',size=20)
plt.text(-2.275,-.024,'o',color='r',size=20)
plt.text(2.195,-.024,'o',color='r',size=20)
plt.text(-2.9,-.15,' Stagnation Point',color='r',size=20)
plt.text(1.75,-.15,' Stagnation Point',color='r',size=20)
plt.text(-1.75,.75,'Contaminated Flow',color='k',size=20)
plt.text(-3.7,.75,'Uncontaminated Flow',color='k',size=20)

plt.clabel(contours, inline=True, fontsize=8)
plt.savefig('rankine.png')
```
