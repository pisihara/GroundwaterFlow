We are given that two monitoring wells MW-1 located at (0,0) with potential $\Phi_1=708.33$ and MW-2 located at  (-4.5,1) with potential $\Phi_2=706.5$ show above action level contaminant, and that three extraction wells EW-1 at (-5,2), EW-2  at (-6.5,1) and EW-3 at (-1,4) are in operation with respective strengths in a 1:5:1:1 ratio.  

We being by assuming the complex potential has the fom

 $\Phi(z)=k_0 x + 1.5 k\ln(\mid z - (-5+2i)\mid)+k\ln(\mid z-(-6.5+i)\mid + k\ln(\mid z-(-1+4i)\mid  + C$.
 
We chose C=712 and estimate $k_0\approx \frac{708.33-706.5}{4.5}\approx .4$, so that 

 $\Phi(z)=.4 x + 1.5 k\ln(\mid z - (-5+2i)\mid)+k\ln(\mid z-(-6.5+i)\mid + k\ln(\mid z-(-1+4i)\mid  + 712$.
 
 

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

Plot of the Equipotential Lines and Streamlines for various $k$.

For $k=.10$ the flow from the two monitoring wells is captured by the middle and lower extraction wells. 

```python
plt.figure(figsize=(20,15))
x = np.linspace(-20, 20, 250)
y = np.linspace(-2.5, 7.5, 250)
plt.xlim(-20, 20)
plt.ylim(-2.5,7.5)
k=.1
k0=.4
c=730
X, Y = np.meshgrid(x, y)
Z = k0*Y+1.5*k*1.5*np.arctan2(Y-2,X+5)+k*np.arctan2(Y-1,(X+6.5))+k*np.arctan2(Y-4,(X+1))
contours=plt.contour(X, Y, Z, levels=np.linspace(-20,20,500), colors='black')
plt.gca().tick_params(labelsize=15)
#plt.clabel(contours, inline=True, fontsize=15)
plt.text(0,0,'oXX',fontsize=20,color='g')
plt.text(-4.5,1,'oXX',fontsize=20,color='g')
plt.savefig('Ex3fig1.png')
```

For a strength of k=.07,there is a streamlines which slips in-between the top and middle extraction wells, and another streamline between the middle and lower extraction wells.  In this case, the lower well captures the streamlines from both monitoring wells.

```python
plt.figure(figsize=(20,15))
x = np.linspace(-20, 20, 250)
y = np.linspace(-2.5, 7.5, 250)
plt.xlim(-20, 20)
plt.ylim(-2.5,7.5)
k=.07
k0=.4
c=730
X, Y = np.meshgrid(x, y)
Z = k0*Y+1.5*k*1.5*np.arctan2(Y-2,X+5)+k*np.arctan2(Y-1,(X+6.5))+k*np.arctan2(Y-4,(X+1))
contours=plt.contour(X, Y, Z, levels=np.linspace(-20,20,500), colors='black')
plt.gca().tick_params(labelsize=15)
#plt.clabel(contours, inline=True, fontsize=15)
plt.text(0,0,'oXX',fontsize=20,color='g')
plt.text(-4.5,1,'oXX',fontsize=20,color='g')
plt.savefig('Ex3fig2.png')
```
