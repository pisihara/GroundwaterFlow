<h4> Introduction</h4>
In this Notebook we consider the problem of an equpotential shoreline represented by a vertical line at $x=5$ with $\Phi= 1$ and an inland pond represented by  circle with radius 2 centered at the origin with a higher potential $\Phi=7.$ Hence, the streamlines flow from the pond to the shore.


We will  use a Mobius Transformation (MT) to obtain the streamlines emanating from the pond towards the shore.  In this case we specify that $f(-2)=-2$, $f(2)=2$ (this is so MT maps the pond boundary to a circle of radius 2 in the $w$ plane) and also that $f(\infty)=-a$ (this is so MT maps shoreline to a concentric circle with radius 1 in the $w$ plane).    That is, we specify that $z_1=-2, z_2=2, z_3=\infty, w_1=-2, w_2=2, w_3=-a$).  The condition $f(\infty)=-a$ wraps the shoreline to a circle which is concentric to the image of the pond boundary circle. This is useful since we know that equipotentials which are concentric circles correspond to a source or sink. The implicit form of Mobius Transformation is given by
      
     
$ (\frac{w+2}{w+a})(\frac{2+a}{4})=(\frac{z+2}{z-\infty})(\frac{z-\infty}{4}) \Rightarrow w=\frac{az-4}{a-z}$.
     
      
   The condition $w=a$ when $z=5$ then gives $a=5+\sqrt{21}$. 
   
   In short, the MT 
   
   $f(z)=\frac{az-4}{a-z}$
   
    with $a=5+\sqrt{21}$ maps the circle of radius 2 in the $z$ plane to the circle of radius 2 in the $w$ plane, and the line $x=5$ to the circle $\mid w \mid=a$.
    
The value $a=5+\sqrt{21}$ can be obtained using $f(a)=5$.
    
The complex potential in the w-plane has the form 
    
$\Omega(w)=c_1\log w + c_2$ 
    
so that 
    
$\Phi(w)=\Phi(r e^{i\theta})=c_1\ln r + c_2$ where 
    
$r=\mid w \mid$. The boundary conditions 

$\Phi(2)=7$ and $\Phi(a)=1$ give 

$c_1\approx -6.54812$ and $c_2\approx 11.53881$.

The potential for the flow in the $z$ plane is therefore $\Omega(f(z))=-6.54812 \log(\frac{az-4}{a-z})$ + 11.53881

From this we obtain the stream function

$\Psi= c_1\arctan (\frac{ay}{ax-4}) + c_1\arctan(\frac{y}{a-x})$.
    

1. Import Libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
import cmath as cm
from time import time
```

2. Plot the Streamlines

```python
# create figure and axes
fig=plt.figure(figsize=(5,5))
# Set up Plot Grid
x = np.linspace(-3, 5, 250)
y = np.linspace(-4, 4, 250)
X, Y = np.meshgrid(x, y)

#Plot the streamlines
plt.gca().tick_params(labelsize=25)
plt.gca().set_title('Streamlines',size=14)
A=5+np.sqrt(21)
C1=-6.54812
Z = C1*np.arctan2(A*Y,A*X-4)+C1*np.arctan2(Y,A-X)
contours=plt.contour(X, Y, Z, levels=np.linspace(-40,40,100), colors='black');
plt.gca().clabel(contours, inline=True, fontsize=8)
plt.savefig('PondToShore.png')
```
