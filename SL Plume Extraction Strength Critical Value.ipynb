This Notebook can be used to plot the Space-Leasing plume potential $\Phi(z)=-.145 x + 3k\ln(\mid z\mid)+2k\ln(\mid z-10.8\mid + 720$ for various choices of the extraction well strength parameter $k$.


In particular, we seek to estimate the critical value $k=k^*$ at which the dividing streamline for the downstream extraction well at (10.8,0) passes through the monitoring well location (-.5,2.5) which has above action-levell contaminant concentration..

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

Plot the Equipotential Lines and Streamlines at the critical value ($k=k^*\approx.35$)

```python
plt.figure(figsize=(20,15))
x = np.linspace(-15, 15, 250)
y = np.linspace(-7, 8, 250)
plt.xlim(-15, 15)
plt.ylim(-7,8)
k=.35
k0=-.145
k1=-.5
k2=.3
k3=.2
c=720
X, Y = np.meshgrid(x, y)
Z = k0*X+.3*k*.5*np.log(Y**2+X**2)+.2*k*.5*np.log(Y**2+(X-10.8)**2)+c
contours=plt.contour(X, Y, Z, levels=np.linspace(700,800,200), colors='red')
Z1 = k0*Y+.3*k*np.arctan2(Y,X)+.2*k*np.arctan2(Y,(X-10.8))
contours1=plt.contour(X, Y, Z1, levels=np.linspace(-10,20,700), colors='black')
plt.gca().tick_params(labelsize=15)
plt.clabel(contours, inline=True, fontsize=15)
plt.text(-.5,2.5,'oXX',fontsize=20,color='g')
plt.savefig('ActionLevel.png')
```
